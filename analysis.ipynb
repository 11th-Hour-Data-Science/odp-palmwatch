### PalmWatch

Investigating the influence of palm oil plantations on Indonesian deforestation

```python
import pandas as pd
from pathlib import Path
import plotly.express as px
```

```python
data = Path("data")
mills = pd.read_csv(data / "mills.csv")
mill_loss = pd.read_csv(data / "mill_loss.csv")
```

```python
mills = mills.rename(columns={"UML Id": "umlid"})
mills_with_loss = pd.concat([mills, mill_loss], axis="columns")
```

```python
mills_with_loss.dropna(subset=["treeloss_2017"], inplace=True)
```

```python
fig = px.bar(mills_with_loss, x="Province", y=["treeloss_2017", "treeloss_2018", "treeloss_2019"], title="Tree Loss by Province, 2017-19")
fig.show()
```
